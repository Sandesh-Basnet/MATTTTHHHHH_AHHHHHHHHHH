## Scatter<br>
Scatter graphs will plot one point per observation, we can use these to graph independent
variables to see if there is a correlation. Python will require two sets of variables for a scatter
graph, one for the x-axis and another for the corresponding y axis. These sets will need to be of
the same length and work best as lists, array, or array-like data types. We can call ‘pyplot’ to plot
a scatter with the pyplot.scatter(x,y) call for example.

Example 1:

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y=[3,4,5,6,7,8,9,10,11]
#plt.plot(x,y)
plt.scatter(x,y,color='red',marker='^')
plt.show()
```

Example 2:

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y=[3,4,5,6,7,1,9,2,5]
#plt.plot(x,y)
plt.scatter(x,y,color='red',marker='^')
x1=[1,5,6,8,12]
y1=[2,6,7,11,13]
plt.scatter(x1,y1,color='blue',marker='s')
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = [random.randint(0, 50) for i in range(10)]
y = [random.randint(0, 50) for i in range(10)]
colors = ["red","green","blue","yellow","pink",
          "orange","purple","brown","cyan","magenta"]
plt.scatter(x, y, c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

**Task 2:** <br>
Following data shows the change in population of different animals from 2017 to 2022.<br>
a. Create a pie chart for the population of each year<br>
b. Can we show all the pie chart on the single pane? Write the code<br>

```python
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 1, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

animals = ["Bears", "Dolphins", "Whales"]

# ---a. Create a pie chart for the population of each year. ---
plt.pie([bears[0], dolphins[0], whales[0]], labels=animals, autopct='%1.1f%%')
plt.title("Animal Population in 2017")
plt.show()

# ---b. Can we show all te pie chart on the single pane? Write the code. ---
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, year in enumerate(years):
    row = i // 3
    col = i % 3
    axes[row, col].pie([bears[i], dolphins[i], whales[i]], labels=animals, autopct='%1.1f%%')
    axes[row, col].set_title(f"Population in {year}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
car_colors = ["Red", "Silver", "Blue", "White", "Green"]
counts = [123, 230, 78, 193, 12]
# --- a) ---
plt.bar(car_colors, counts)
plt.title("Car Colors Seen")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()
# --- b) ---
plt.bar(car_colors, counts, width=0.3)
plt.title("Car Colors Seen (Narrow Bars)")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()
# --- c) ---
plt.barh(car_colors, counts)
plt.title("Car Colors Seen (Horizontal)")
plt.xlabel("Number Seen")
plt.ylabel("Car Color")
plt.show()
# --- d) ---
mycolors = ["red", "silver", "blue", "white", "green"]
plt.bar(car_colors, counts, color=mycolors)
plt.title("Car Colors Seen (Colored Bars)")
plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.show()
# --- e) ---
plt.pie(counts, labels=car_colors, autopct='%1.1f%%')
plt.title("Car Colors Distribution")
plt.show()
# --- f) ---
# A: Explode one slice (Red cars)
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(counts, labels=car_colors, autopct='%1.1f%%', explode=myexplode)
plt.title("Car Colors Distribution (Exploded Red)")
plt.show()
# B: Explode + Shadow
plt.pie(counts, labels=car_colors, autopct='%1.1f%%', explode=myexplode, shadow=True)
plt.title("Car Colors Distribution (Exploded + Shadow)")
plt.show()
# C: Add Legend
plt.pie(counts, labels=car_colors, autopct='%1.1f%%')
plt.legend(title="Car Colours:")
plt.title("Car Colors Distribution (With Legend)")
plt.show()
# D: Custom Colours
mycolours = ["magenta", "green", "cyan", "red", "yellow"]
plt.pie(counts, labels=car_colors, autopct='%1.1f%%', colors=mycolours)
plt.title("Car Colors Distribution (Custom Colours)")
plt.show()
```
