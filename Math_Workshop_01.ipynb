1. Write a program to prompt the user for hours and rate per hour using input( ) to compute gross pay. Pay the hourly rate for the hours up to 40 and 1.5 times the hourly rate for all hours worked above 40 hours. Use 45 hours and a rate of 10.50 per hour to test the program (the pay should be 498.75).

```python
hrs = input("Enter Hours: ")
h = float(hrs)
rate = input("Enter the rate per hour: ")
r = float(rate)
if h <= 40:
    pay = h * r
else:
    overtime = h - 40
    pay = (40 * r) + (overtime * r * 1.5)
print("Gross Pay:", pay)
```

### Write a function fib (n) that prompts the number of terms “n” from the user and displays first “n” terms of Fibonacci sequence iteratively

```python
def fibo(n_iterations):
    a = 0
    b = 1
    count = 0
    print("Fibonacci Series:")

    while count < n_iterations:
        print(a)
        next_num = a + b
        a = b
        b = next_num
        count += 1
n = int(input("Enter the number of iteration: "))
fibo(n)
```

### Write a function myPow(x, n) that returns 𝑥𝑛, where and n is a non-negative integer. Do not use the ** operator or the math.pow function - use one while loop.

```python
def myPow(x, n):
    result = 1
    i = 0
    while i < n:
        result = result * x
        i += 1
    return result
x = int(input("Enter number of terms: "))
n = int(input("Enter number of terms: "))
print(myPow(x, n))
```

### Write a function SumDigits(n) that displays the sum of digits of “n”. Hint: SumDigits(234) will be 2+3+4=9

```python
def SumDigits(number):
    str_number = str(number)
    total = 0
    for i in str_number:
        total += int(i)
    print(total)

SumDigits(234)
```

### Write a function squaredSum that returns the sum of the squares of an array/list of integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python to show the logic in your solution, do not use any library functions.

```python
def squaredSum(arrryyyy):
  total = 0
  for i in range(len(arrryyyy)):
    arrryyyy[i] = arrryyyy[i]**2
    total += arrryyyy[i]
  print(total)
squaredSum([2,3,4])
```

### Write a function called sumDigits that is given an integer num and returns the sum of the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if there is one)
